```python
day=0
time=0
while day<3:
    day+=1
    print('{}일 입니다. 좋은 하루 되세요!'.format(day))
    while time<3:
        time+=1
        if time==1:
            print('아침 시간입니다. 아침 드세요~')
        elif time==2:
            print('점심 배부르게 드세요!')
        else:
            print('나는 밥 대신 술~')
    time=0
```

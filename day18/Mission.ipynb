```python
from matplotlib import font_manager,rc
import matplotlib.pyplot as plt
font_path='C:/Windows/Fonts/malgunsl.ttf'
font_name=font_manager.FontProperties(fname=font_path).get_name()
plt.rc('font',family=font_name)
figure=plt.figure()
axes=figure.add_subplot(111)
ratio = [0.18,0.3,3.33,3.75,0.38,25,0.25,2.75,0.1]
labels = ['비타민C','비타민 B1','비타민 B2','나이아신','비타민 B6','비타민 C','비타민 D','비타민 E','엽산']
plt.pie(ratio, labels=labels, autopct='%.2f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random as rd
figure=plt.figure()
axes=figure.add_subplot(111)
x1=[]
y1=[]
for i in range(1,101):
    i=rd.randint(1,101)
    x1.append(i)
for j in range(1,101):
    j=rd.randint(1,101)
    y1.append(j)    
axes.bar(x1,y1,color='green')
x2=[]
y2=[]
for i in range(1,101):
    x2.append(i)
for j in range(1,101):
    j=rd.randint(1,101)
    y2.append(j)
axes.plot(x2,y2,linestyle='-',linewidth='2.0',marker='.',color='red')
plt.show()
```
